```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
scores = []
for i in range(1, 6):
    data = pickle.load(open(f"Checkpoints/total_score_CartPole-v1_1_{i}.pkl", "rb"))
    scores.append(data)

scores = np.array(scores)

transposed_scores = np.transpose(scores)
scores = transposed_scores
mean_scores1 = np.mean(scores, axis=1)

scores = []
for i in range(1, 6):
    data = pickle.load(open(f"Checkpoints/total_score_CartPole-v1_2_{i}.pkl", "rb"))
    scores.append(data)

scores2 = np.array(scores)

transposed_scores = np.transpose(scores2)
scores2 = transposed_scores
mean_scores2 = np.mean(scores2, axis=1)

plt.plot(mean_scores1, label="Type-1")
plt.plot(mean_scores2, label="Type-2")
plt.title("Mean scores for CartPole-v1")

plt.xlabel("Episode")
plt.ylabel("Mean score")
plt.legend()
plt.show()
```
